# Data Wrangling

```python
# 1. Aadir el newspaper_uid al DataFrame
el_tiempo['newspaper_uid'] = 'eltiempo'
el_tiempo
```

```python
# Obtener el host
from urllib.parse import urlparse
el_tiempo['host'] = el_tiempo['url'].apply(lambda url: urlparse(url).netloc)
el_tiempo
```

```python
# Obtener cuantos hosts se repiten dentro del dataframe
el_tiempo['host'].value_counts()
```

```python
missing_titles_mask = el_tiempo['title'].isna()
missing_titles_mask
```

```python
missing_titles = (el_tiempo[missing_titles_mask]['url']
                    .str.extract(r'(?P<missing_titles>[^/]+)$')
                    .astype(str).applymap(lambda title: title.replace('-', ' '))
                    #.applymap(lambda title: title_word_list: ' '.join(title_word_list) )
                    )
missing_titles
```

# Additional CleanUp

```python
# 4. Aadir UID a las Filas
import hashlib
# axis eje 1 significa las filas eje 0 columnas
uids = (el_tiempo
            .apply(lambda row: hashlib.md5(bytes(row['url'].encode())) , axis=1)
            .apply(lambda hash_object: hash_object.hexdigest())
        )
el_tiempo["UIDS"] = uids
el_tiempo.set_index('UIDS', inplace=True)
el_tiempo
```

```python
stripped_body = (el_tiempo
                    .apply(lambda row: row['body'], axis=1)
                    .apply(lambda body: body.replace('\n',''))
                    .apply(lambda body: body.replace('\r',''))
                )
stripped_body
```

# Data enrichment

```python
# Tokenizar el titulo y el body
import nltk
from nltk.corpus import stopwords
# nltk.download('punkt')
# nltk.download('stopwords')
```

```python
stopwords = set(stopwords.words('spanish'))
```

```python
def tokenize_colum(df, column_name):
    return (df
            .dropna()
            .apply(lambda row: nltk.word_tokenize(row[column_name]), axis=1)
            .apply(lambda tokens: list(filter(lambda token: token.isalpha(), tokens)))
            .apply(lambda tokens: list(map(lambda token: token.lower(), tokens)))
            .apply(lambda word_list: list(filter(lambda word: word not in stopwords, word_list)))
            .apply(lambda valid_word_list: len(valid_word_list))
            )

el_tiempo['n_token_body'] = tokenize_colum(el_tiempo, 'body')
el_tiempo
```
